```python
import requests as req
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import matplotlib
```

```python
try:
    html = req.get("https://www.worldometers.info/coronavirus", timeout=5)
except req.exceptions.Timeout:
    print("Error: The request took too long.")
except req.exceptions.RequestException as e:
    print(f"Network or HTTP error:{e}")
```

```python
html.content
```

```python
html_parsed = BeautifulSoup(html.content)
```

```python
table = html_parsed.find("table", attrs = {"id": "main_table_countries_today"})
```

```python
table
```

```python
rows = table.find_all("tr")
```

```python
rows[0]
```

```python
rows[0].text.strip()
```

```python
rows[9].text.strip().split("\n")
```

```python
data = []
for r in rows:
    data.append(r.text.strip().split("\n")[1:5])
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df = pd.DataFrame(data[9:], columns = data[0])
```

```python
df.head()
```

```python
df.to_csv("covid19.csv")
```

```python
df_plot = df[["Country,Other", "TotalCases"]]
df_plot = df_plot[:10]
```

```python
df_plot.head()
```

```python
df_plot["TotalCases"] = df_plot["TotalCases"].apply(lambda x: x.replace(",", "")).astype(int)
```

```python
df_plot.head()
```

```python
df_plot.plot(kind = "bar", x = "Country,Other", y = "TotalCases")
```
